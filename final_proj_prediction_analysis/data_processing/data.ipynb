```python
df = merged_df.copy()
df["median_income_2017"] = pd.to_numeric(df["median_income_2017"], errors="coerce")
df["median_income_2019"] = pd.to_numeric(df["median_income_2019"], errors="coerce")
df['income_growth'] = (df['median_income_2019'] - df['median_income_2017']) / df['median_income_2017']
baseline_median_income = df['median_income_2017'].median()
median_growth = df['income_growth'].median()

def label_neighborhood(row, income_thresh, growth_thresh):
    if row['median_income_2017'] >= income_thresh:
        return "Higher-Income"
    else:
        if row['income_growth'] > growth_thresh:
            return "Gentrifying"
        else:
            return "Non-Gentrifying"

df['gentrification_label'] = df.apply(
    label_neighborhood,
    axis=1,
    args=(baseline_median_income, median_growth)
)

print(df[['Sub-Borough Area', 'median_income_2017', 'median_income_2019',
          'income_growth', 'gentrification_label']].head(15))


df.to_csv("final_gentrification_dataset_with_labels.csv", index=False)
```
