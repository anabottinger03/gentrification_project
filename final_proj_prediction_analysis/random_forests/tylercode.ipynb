## Decision Tree

```python
import numpy as np

def entropy(y):
    """
    entropy of a label distribution
    y: array-like of labels.
    """
    # Get unique labels and their counts
    values, counts = np.unique(y, return_counts=True)
    probs = counts / len(y)
    # Compute entropy (using base-2 logarithm)
    return -np.sum(probs * np.log2(probs + 1e-9))  # add a tiny constant to avoid log(0)

def information_gain(y, y_left, y_right):
    """
    information gain from splitting a dataset into y_left and y_right
    """
    H = entropy(y)
    H_left = entropy(y_left)
    H_right = entropy(y_right)
    p_left = len(y_left) / len(y)
    p_right = len(y_right) / len(y)
    return H - (p_left * H_left + p_right * H_right)

class DecisionTreeNode:
    def __init__(self, feature_index=None, threshold=None, left=None, right=None, *, value=None):
        """
        represents a decision tree node
        """
        self.feature_index = feature_index  # index of the feature to split on
        self.threshold = threshold          # threshold value for the split
        self.left = left                    # left child node
        self.right = right                  # right child node
        self.value = value                  # predicted class at leaf

class DecisionTreeClassifier:
    def __init__(self, max_depth=5, min_samples_split=2):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.root = None
    
    def fit(self, X, y):
        """
        Build the decision tree by recursively splitting the data.
        X: 2D numpy array of features.
        y: 1D numpy array of labels.
        """
        self.root = self._build_tree(X, y, depth=0)
    
    def _build_tree(self, X, y, depth):
        n_samples, n_features = X.shape
        
        # Stopping criteria: pure node, maximum depth, or too few samples
        if len(np.unique(y)) == 1 or depth >= self.max_depth or n_samples < self.min_samples_split:
            leaf_value = self._most_common_label(y)
            return DecisionTreeNode(value=leaf_value)
        
        best_gain = -1
        best_feature = None
        best_threshold = None
        best_left_indices = None
        best_right_indices = None
        
        # Iterate over all features
        for feature_index in range(n_features):
            X_feature = X[:, feature_index]
            thresholds = np.unique(X_feature)
            for threshold in thresholds:
                left_indices = np.where(X_feature <= threshold)[0]
                right_indices = np.where(X_feature > threshold)[0]
                if len(left_indices) == 0 or len(right_indices) == 0:
                    continue
                y_left = y[left_indices]
                y_right = y[right_indices]
                gain = information_gain(y, y_left, y_right)
                if gain > best_gain:
                    best_gain = gain
                    best_feature = feature_index
                    best_threshold = threshold
                    best_left_indices = left_indices
                    best_right_indices = right_indices
        
        # if no split, make a leaf
        if best_gain == -1:
            leaf_value = self._most_common_label(y)
            return DecisionTreeNode(value=leaf_value)
        
        # build left and right subtree
        left_subtree = self._build_tree(X[best_left_indices, :], y[best_left_indices], depth+1)
        right_subtree = self._build_tree(X[best_right_indices, :], y[best_right_indices], depth+1)
        return DecisionTreeNode(feature_index=best_feature, threshold=best_threshold, 
                                left=left_subtree, right=right_subtree)
    
    def _most_common_label(self, y):
        """
        Return the most common label in y.
        """
        values, counts = np.unique(y, return_counts=True)
        return values[np.argmax(counts)]
    
    def predict(self, X):
        """
        Predict class labels for samples in X.
        """
        return np.array([self._predict_sample(x, self.root) for x in X])
    
    def _predict_sample(self, x, node):
        """
        Recursively traverse the tree to get a prediction for a single sample.
        """
        if node.value is not None:
            return node.value
        if x[node.feature_index] <= node.threshold:
            return self._predict_sample(x, node.left)
        else:
            return self._predict_sample(x, node.right)
```

## Random Forest 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc

# helper functions

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def logistic_loss(y, p):
    """
    Compute average logistic loss.
    y: true labels (0 or 1)
    p: predicted probabilities (should be in (0,1))
    """
    # Add a tiny constant to avoid log(0)
    eps = 1e-9
    return -np.mean(y * np.log(p + eps) + (1 - y) * np.log(1 - p + eps))

# regression tree 

class RegressionTreeNode:
    def __init__(self, feature_index=None, threshold=None, left=None, right=None, value=None):
        self.feature_index = feature_index  
        self.threshold = threshold          
        self.left = left                    
        self.right = right                  
        self.value = value                  

class RegressionTree:
    def __init__(self, max_depth=3, min_samples_split=2):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.root = None
    
    def fit(self, X, y):
        self.root = self._build_tree(X, y, depth=0)
    
    def _build_tree(self, X, y, depth):
        n_samples, n_features = X.shape
        # Stopping conditions
        if depth >= self.max_depth or n_samples < self.min_samples_split or np.unique(y).size == 1:
            return RegressionTreeNode(value=np.mean(y))
        
        best_feature = None
        best_threshold = None
        best_loss = np.inf
        best_left_indices = None
        best_right_indices = None
        
        # Try all splits
        for feature_index in range(n_features):
            X_feature = X[:, feature_index]
            thresholds = np.unique(X_feature)
            for threshold in thresholds:
                left_indices = np.where(X_feature <= threshold)[0]
                right_indices = np.where(X_feature > threshold)[0]
                if len(left_indices) == 0 or len(right_indices) == 0:
                    continue
                left_y = y[left_indices]
                right_y = y[right_indices]
                loss = (len(left_y)/n_samples) * np.var(left_y) + (len(right_y)/n_samples) * np.var(right_y)
                if loss < best_loss:
                    best_loss = loss
                    best_feature = feature_index
                    best_threshold = threshold
                    best_left_indices = left_indices
                    best_right_indices = right_indices
        
        if best_feature is None:
            return RegressionTreeNode(value=np.mean(y))
        
        left_subtree = self._build_tree(X[best_left_indices, :], y[best_left_indices], depth + 1)
        right_subtree = self._build_tree(X[best_right_indices, :], y[best_right_indices], depth + 1)
        return RegressionTreeNode(feature_index=best_feature, threshold=best_threshold,
                                  left=left_subtree, right=right_subtree)
    
    def _predict_sample(self, x, node):
        if node.value is not None:
            return node.value
        if x[node.feature_index] <= node.threshold:
            return self._predict_sample(x, node.left)
        else:
            return self._predict_sample(x, node.right)
    
    def predict(self, X):
        return np.array([self._predict_sample(x, self.root) for x in X])


   
    

```

## Gradient boosting

```python
class GradientBoostingClassifier:
    def __init__(self, n_estimators=10, learning_rate=0.1, max_depth=3, min_samples_split=2):
        if n_estimators < 1:
            raise ValueError("n_estimators must be positive")
        if learning_rate <= 0:
            raise ValueError("learning_rate must be positive")
        if max_depth < 1:
            raise ValueError("max_depth must be positive")
        if min_samples_split < 2:
            raise ValueError("min_samples_split must be at least 2")
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.trees = []
        self.F0 = None
        self.train_loss = []
        self.val_loss = []

    def fit(self, X, y, X_val=None, y_val=None, early_stopping_rounds=None):
        X = np.asarray(X)
        y = np.asarray(y)
        if X.ndim != 2:
            raise ValueError("X must be a 2D array")
        if y.ndim != 1:
            raise ValueError("y must be a 1D array")
        if X.shape[0] != y.shape[0]:
            raise ValueError("X and y must have the same number of samples")
        if not np.all(np.isin(y, [0, 1])):
            raise ValueError("y must contain only 0 or 1 for binary classification")
        if len(np.unique(y)) < 2:
            raise ValueError("y must contain both classes (0 and 1)")

        p = np.mean(y)
        p = np.clip(p, 1e-15, 1 - 1e-15)
        self.F0 = np.log(p / (1 - p))
        F = np.full(shape=y.shape, fill_value=self.F0)

        for m in range(self.n_estimators):
            p_current = sigmoid(F)
            residuals = y - p_current
            tree = RegressionTree(max_depth=self.max_depth, min_samples_split=self.min_samples_split)
            tree.fit(X, residuals)
            self.trees.append(tree)
            F += self.learning_rate * tree.predict(X)
            loss = logistic_loss(y, sigmoid(F))
            self.train_loss.append(loss)

            if X_val is not None and y_val is not None:
                p_val = sigmoid(self._compute_F(X_val))
                val_loss = logistic_loss(y_val, p_val)
                self.val_loss.append(val_loss)
                if early_stopping_rounds and m > early_stopping_rounds:
                    if val_loss >= min(self.val_loss[-early_stopping_rounds:]):
                        break

    def _compute_F(self, X):
        F = np.full(shape=(X.shape[0],), fill_value=self.F0)
        for tree in self.trees:
            F += self.learning_rate * tree.predict(X)
        return F

    def predict_proba(self, X):
        F = self._compute_F(X)
        return sigmoid(F)

    def predict(self, X):
        proba = self.predict_proba(X)
        return (proba >= 0.5).astype(int)
    

# feature importance
def get_feature_usage(node, feature_counts):
    if node is None:
        return
    # If leaf, return
    if node.value is not None:
        return
    # Increase count for the feature used at this node
    feature_counts[node.feature_index] += 1
    get_feature_usage(node.left, feature_counts)
    get_feature_usage(node.right, feature_counts)

def compute_feature_importances(trees, n_features):
    feature_counts = np.zeros(n_features)
    for tree in trees:
        get_feature_usage(tree.root, feature_counts)
    total = np.sum(feature_counts)
    if total > 0:
        return feature_counts / total
    else:
        return feature_counts
```
